```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
#https://data.wprdc.org/datastore/dump/513290a6-2bac-4e41-8029-354cbda6a7b7
waterFeatures = pd.read_csv("https://data.wprdc.org/datastore/dump/513290a6-2bac-4e41-8029-354cbda6a7b7")
```

```python
query_mask = waterFeatures["feature_type"] == "Spray"
waterParks = waterFeatures[query_mask]
waterParks.head(10)
```

```python
waterPark = waterParks['neighborhood'].value_counts()
waterPark.plot.bar(stacked=True, title="Water Parks")
```

## Grass

```python
#https://data.wprdc.org/dataset/operations-green-spaces/resource/b06fa202-ead7-491c-b8f6-2702dd35f8dd
grass = pd.read_csv("opendata-pubworks-operations-green-spaces-2015.csv")
grass.head(10)
```

```python
query_mask = grass["Total Acreage"] >= .1
grassSize = grass[query_mask]
grasses = grassSize["Neighborhood"].value_counts()
grasses.plot.bar(stacked=True, title="Green Areas Larger than .1 Acres")
```
