I will first try using the signalized intersecion datasets. Since Panthers do not like to stop while running, so they would prefer the neighborhood with the least amount of traffic signs that slow them down.

I will look at the traffic sign data base at https://data.wprdc.org/dataset/city-traffic-signs.
This dataset contains the traffic signs that are in Pittsburgh. It is frequently updated since the dataset was initially posted to the WPRDC in 2017 but was last updated three days ago.

```python
cityTrafficSignals = pd.read_csv("https://data.wprdc.org/datastore/dump/d078a6b5-83a3-4723-a3a9-5371cfe1cc0c")
```

Lets print out the dataset to see how it is formatted:

```python
cityTrafficSignals
```

I will group the dataset according to neighborhood and count totals of all the columns to see if the fields properly count the number of traffic signs.

```python
cityTrafficSignals.groupby("neighborhood").count()
```

Compared to the find function in a spreadsheet program, the description field accounts for each traffic sign in a given neighborhood when using count().

When going through all the types of signs within the dataset the most prominent ones which would affect the Panthers in there running are the ones that make them slow down or stop. Since the Stop sign and Yield sign are the two types of signs which would result in slowing down the panthers, I will take query masks of the stop signs and yield signs individually.

```python
# Get query masks of stp signs and yield signs
stopSigns = cityTrafficSignals['description'] == "Stop"
yieldSigns = cityTrafficSignals['description'] == "Yield"
```

I will create two dataframes one with only stop signs and the other with only yield signs.

```python
#Dataframe with only stop signs
trafficStops = cityTrafficSignals[stopSigns]
#Dataframe with only yield signs
trafficYields = cityTrafficSignals[yieldSigns] 
```

I will take the same approach as above and take the two individual dataframes and get the count of the signs per neighborhood in a series.

```python
tSN = trafficStops.groupby("neighborhood").count()["description"]
tYN = trafficYields.groupby("neighborhood").count()["description"]
```

Lets print out the series and describe them to see the data.

```python
print("There are " + str(tSN.sum()) + " stop signs recorded in this dataset")
print("There are " + str(tYN.sum()) + " yield signs recorded in this dataset")
```

There are fewer records of yield signs in the dataseat as compared to stop signs. To remove the effect of the minimally recorded yield signs, I will give less weight to the yield signs in comparison to the stop signs. Additionally, Panthers are not as drastically effected by yield signs as compared to stop signs, since they do not have to stop if there are no other Panthers going on the road intersecting with the one they are on.

```python
#Weighting the Signs
WTrafficYields = tYN*0.4
WTrafficStops = tSN*0.6
```

To compare the stop signs and yield signs in each neighborhood, I will create a new data frame including the weighted stop signs and yield signs with the totals for each neighborhood.

```python
trafficSigns = pd.concat([WTrafficStops.rename("Stops"), WTrafficYields.rename("Yields")], axis=1).fillna(0)
#Adding the sum of the yields and stops in a column named total
trafficSigns["Total"] = trafficSigns["Stops"] + trafficSigns["Yields"]
```

```python
trafficSigns
```

```python
# Here I am sorting the data according to its total in decending order
trafficSigns.sort_values('Total', inplace=True, ascending=False)
```

```python
# Here I am creating a stacked bar chart containg the weighted stop signs and weighted yield signs
trafficSigns[['Stops', 'Yields']].plot(kind='barh', stacked=True, figsize=(30, 30), fontsize=20)
plt.title("Neighborhoods Vs. Signs", fontsize=20)
plt.xlabel("Stops + Yield Signs", fontsize=20)
plt.ylabel("Neighborhoods", fontsize=20)
plt.legend(prop={'size': 40})
```

```python
#Get shapefile for Pittsburgh neighborhoods
neighborhoods = geopandas.read_file("../Neighborhoods/Neighborhoods_.shp")
# Merge the neighbor hood data with the traffic signs data
sign_map = neighborhoods.merge(trafficSigns[['Total']], how='left', left_on='hood', right_on='neighborhood')
```

```python
# Here I print out the geopandas plot
sign_map.plot(column='Total', # set the data to be used for coloring
               cmap='OrRd',              # choose a color palette
               edgecolor="white",        # outline the districts in white
               legend=True,              # show the legend
               legend_kwds={'label': "Weighted Traffic Sign Total"}, # label the legend
               figsize=(15, 10),         # set the size
               )
```
